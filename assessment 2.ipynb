```python
matrix1 = []
print("Enter values of matrix1:")
for i in range(3):
    row = []
    for j in range(2):
        row.append(float(input(f"Enter elements ")))
    matrix1.append(row)


matrix2 = []
print("Enter values of matrix2 ")
for i in range(2):
    row = []
    for j in range(3):
        row.append(float(input(f"Enter elements")))
    matrix2.append(row)


result_matrix = [[0 for _ in range(3)] for _ in range(3)]


for i in range(3):
    for j in range(3):
        for k in range(2):
            result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

print("\nResulting Matrix")
for row in result_matrix:
    print(row)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

data=pd.read_csv('/content/drive/MyDrive/assessment.txt')
x=np.array(data).reshape((-1,1))
y=np.array(data)

plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='blue',marker='*')
print("MSE:",mse)
print("intercept",SLR.intercept_)
print("slope",SLR.coef_)
```
