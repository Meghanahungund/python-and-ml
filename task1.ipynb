1. Read a CSV file with data in multiple columns and compute mean, median mode of each of the columns.

```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/items12 - Sheet1.csv')
print(data)
mean1 = data['quantity'].mean()
print('mean of quantity: ' ,mean1)
mean2 = data['unitcost'].mean()
print('mean of unitcost: ' ,mean2)
mean3 = data['totalcost'].mean()
print('mean of totalcost: ', mean3)
median1 = data['quantity'].median()
print('median of quantity: ', median1)
median2 = data['unitcost'].median()
print('median of unitcost: ', median2)
median3 = data['totalcost'].median()
print('median of totalcost: ', median3)
mode1 = data['quantity'].mode()
print('mode of quantity: ', mode1)
mode2 = data['unitcost'].mode()
print('mode of unitcost: ', mode2)
mode3 = data['totalcost'].mode()
print('mode of totalcost: ', mode3)

```

2. Plot each of the columns with different colours (check what kind of graph is suitable )



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('/content/drive/MyDrive/items12 - Sheet1.csv')
print(data)
x=np.array(data['items'])
y=np.array(data['quantity'])
plt.bar(x,y,color='b')
plt.title('quantity',fontsize = 10)
plt.xlabel('items')
plt.ylabel('quantity')
plt.show()
x=np.array(data['items'])
y=np.array(data['unitcost'])
plt.bar(x,y,color='r')
plt.title('unitcost',fontsize = 10)
plt.xlabel('items')
plt.ylabel('unitcost')
plt.show()
x=np.array(data['items'])
y=np.array(data['totalcost'])
plt.bar(x,y,color='g')
plt.title('totalcost',fontsize = 10)
plt.xlabel('items')
plt.ylabel('totalcost')
plt.show()
```

3. Create a game (like lottery/ rock paper/scissors) and write a python code to record how many times the computer won and the user won in an excel file. Plot the graph showing the success rate of user and computer with different colours.

```python
import random
import matplotlib.pyplot as plt
import numpy as np
a=0
b=0
total_guesses = 7
print("welcome to guessing game")
for i in range(total_guesses):
 computer=random.randint(1,10)
 print("")
 user=int(input("guess the number between  1 and 10"))
 print(computer)
 if computer==user:
  print("congraluations you won ")
  a=a+1
 elif computer!=user:
  print("computer won")
  b=b+1
 elif (i == total_guesses - 1):
  print("Sorry, you have run out of guesses. You lose!")
print(" In 7 chances computer wins ",b ,"times")
print(" In 7 chances user wins" ,a,"times")
x=np.array(["user","computer"])
y=np.array([a,b])
plt.pie(y,labels=["user","computer"])
plt.title('sucess rate',fontsize = 15)
plt.show()

```
