args and kwargs

```python
def add(*args):
  sum=0
  for i in args:
    sum+=i
  return sum
a=add(1,2,5,8)
print(a)
```

```python
def greet(**kwargs):
 print("hello")
 for key,value in kwargs.items():
  print(value)
greet(name="meghana",usn="01fe21bec213",branch="ec")
```

rock paper scissor


```python
import random
gamelist=["rock","paper","scissor"]
z=random.choice(gamelist)
user=input("enter either rock or paper or scissor")
print(z)
print(user)
if(z=="rock"and user=="paper"):
 print("user won")
elif(z=="paper"and user=="scissor"):
  print("user won")
elif(z=="rock"and user=="scissor"):
  print("z won")
elif(z=="scissor"and user =="paper"):
  print("z won")
else:
  print("tie")
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')
x=np.linspace(0,10,100)
y=6+3*np.sin(2*x)


fig, a=plt.subplots()
a.plot(x,y,linewidth=3.0)

a.set(xlim=(0,8),xticks=np.arange(1,8),
      ylim=(0,8),yticks=np.arange(1,8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
temp=np.array([10,20,25,30,35,40,50,60,70,80,90])
res=np.array([3453,353,3535,5534,3455,3545,433,434,788,8989,8776])
plt.title("thermistor chacteristics")
plt.xlabel("temp")
plt.ylabel("res")
plt.scatter(temp,res)
plt.grid()
plt.show()
```
